```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('arrival_rate.csv')

# Plot Avg Response Time vs Arrival Rate
df.plot(x="Arrival Rate", y="Avg Response Time (s)", kind="line", title="Response Time vs Arrival Rate")
plt.xlabel("Arrival Rate (requests/second)")
plt.ylabel("Average Response Time (s)")
plt.show()

# Plot Failed Requests vs Arrival Rate
df.plot(x="Arrival Rate", y="Failed Requests", kind="line", title="Failed Requests vs Arrival Rate")
plt.xlabel("Arrival Rate (requests/second)")
plt.ylabel("Failed Requests")
plt.show()
```
